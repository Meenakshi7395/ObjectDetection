```python
import numpy as np
import matplotlib.pyplot as plt

# यह कोड आपके लैपटॉप में एक 3D स्क्रीन खोलेगा जो मशीन का वर्किंग मॉडल दिखाएगी
def run_electron_beam_simulation():
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # 1. इलेक्ट्रॉन सोर्स (Gun) - सबसे ऊपर
    ax.scatter([0], [0], [10], color='red', s=200, label='Electron Gun (Source)')
    
    # 2. मोटी इलेक्ट्रॉन बीम (Main Beam)
    z_main = np.linspace(7, 10, 20)
    x_main = np.zeros(20)
    y_main = np.zeros(20)
    ax.plot(x_main, y_main, z_main, color='orange', linewidth=4, label='Main Electron Beam')
    
    # 3. अपेर्चर एरे (चलनी / Plate) - बीच में
    ax.plot_surface(np.array([[-1, 1], [-1, 1]]), np.array([[-1, -1], [1, 1]]), np.array([[7, 7], [7, 7]]), color='gray', alpha=0.5)
    ax.text(1.2, 0, 7, "Aperture Array (Plate)", color='black')
    
    # 4. 1 बीम से बंटी 4 बारीक बीम्स (Multi-Beams)
    beam_offsets = [(-0.5, -0.5), (-0.5, 0.5), (0.5, -0.5), (0.5, 0.5)]
    colors = ['cyan', 'blue', 'green', 'magenta']
    
    for i, (offset_x, offset_y) in enumerate(beam_offsets):
        z_sub = np.linspace(0, 7, 50)
        x_sub = np.full(50, offset_x)
        y_sub = np.full(50, offset_y)
        ax.plot(x_sub, y_sub, z_sub, color=colors[i], linestyle='--', linewidth=1.5)
        
        # वेफर पर पड़ने वाले नैनो-डॉट्स (1nm Circuit Printing)
        ax.scatter([offset_x], [offset_y], [0], color='gold', s=50)
    
    # 5. सिलिकॉन वेफर (Silicon Wafer) - सबसे नीचे
    ax.plot_surface(np.array([[-2, 2], [-2, 2]]), np.array([[-2, -2], [2, 2]]), np.array([[0, 0], [0, 0]]), color='darkblue', alpha=0.3)
    ax.text(2.2, 0, 0, "Silicon Wafer (1nm Output)", color='darkblue', weight='bold')
    
    # टूल सेटिंग्स
    ax.set_xlabel('X Axis (Position)')
    ax.set_ylabel('Y Axis (Position)')
    ax.set_zlabel('Z Axis (Machine Height)')
    ax.set_title('Stoic Multi-Beam Electron Lithography 3D Simulation')
    ax.legend(loc='upper left')
    
    plt.show()

if __name__ == "__main__":
    run_electron_beam_simulation()
```
